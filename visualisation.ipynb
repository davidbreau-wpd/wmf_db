```python
import matplotlib.pyplot as plt, pandas as pd, seaborn as sns; sns.set_theme(style="whitegrid")
from pathlib import Path
```

```python
df = pd.read_csv(Path('data') / 'Database_inoperation.csv')
```

```python
supplier_counts = df["WEC service company"].value_counts() # occurrences de chaque fournisseur

#Graphique
plt.figure(figsize=(8, 8))              # Taille de l'image
plt.pie(supplier_counts,                # Graphique circulaire des occurences
        labels=supplier_counts.index,   # Etiquettes
        autopct='%1.1f%%',              # Affiche les pourcentages 
        startangle=300,                 # Inclinaison
        colors=sns.color_palette("Set3", len(supplier_counts)))

# Titres et ajustements
plt.title("Répartition des parcs en activité par supplier", fontsize=16)

# Afficher le graphique
plt.axis('equal')  # Pour que le graphique soit un cercle
plt.show()
```
